## The Effect of Depth on Blue Light Irradiance and the Implications for *Nereocystis luetkeana* Recruitment


### Introduction

The life history theory states there are patterns of biological traits during the life of an organism and these traits are shaped by natural selection to optimize survival and reproduction (Petrik 2019). Dormancy is a common life history trait of many microbes, plants, and animals and can be described as a spectrum of inactive states that permit an organisms’ ability to survive periods of unsuitable conditions in variable environments (Evans et al. 2005, Agrawal et al. 2021, Miller et al. 2021). In terrestrial plants, dormancy is a state where seeds or other tissues reduce metabolic activity and have no growth during stressful conditions, such as winter or drought (Bewley and Black 1994, Miller et al. 2021). In many algal species, there is a lag time between the release of zoospores and the appearance of juvenile sporophytes in the field (Hoffman and Santelices 1991, Veenhof et al. 2022). These lag times are often related with climatic conditions, such as light, temperature, and storms during the winter. During these periods, species can persist through microscopic phases that are able to withstand those climatic conditions that would be stressful for macroscopic thalli (Hoffman and Santelices 1991). The persistence of these microscopic stages can be characterized as dormancy and is attributed to the ability to remain in a vegetative state or the ability to delay development of the next developmental stage (Gaitán-Espitia et al. 2014, Edwards 2022, Veenhof et al. 2022). 

*Nereocystis luetkeana*, commonly known as bull kelp, is an annual species that grows from >30m deep along rocky coastlines (Springer et al. 2006). Since it is an annual species, it completes its life cycle in one year which means the survival of microscopic stages is integral to the persistence of the population after it completes its life cycle (Carney and Edwards 2010). The species exhibit a heteromorphic life cycle where there are alternate stages of macroscopic and microscopic structures (fig. 1). During bull kelp’s reproductive season, in the fall, sori form on surface blades and drop to the benthos where zoospores are released and develop into gametophytes. During the winter, gametophytes exist in a vegetative state until conditions become favorable for sporophyte growth in the spring (Lind and Konar 2017). Gametophytes are microscopic structures that develop male and female reproductive organs known as the antheridium and the oogonium which produce gametes for fertilization through a process called gametogenesis. The gametophytes will remain in this vegetative state until certain environmental cues occur and cause a phase shift from dormancy to gametogenesis. The initiation of gametogenesis is caused by photomorphogenetic reactions where a sensory pigment senses light qualities and causes changes to the structural development (Lüning 1980, Lüning 1990). For instance, in aquaculture, it was proven that red or green light causes gametophytes to grow vegetatively while blue light induces gametogenesis (Lüning and Dring 1972, Lüning and Dring 1975, Lüning and Neushul 1978) which may be related to the natural light attenuation as depth increases in marine environments which controls kelp forest growth ranges (Schoenrock et al. 2020).

In the field, different wavelengths of solar radiation, such as red, infrared, and blue, attenuate at different depths (Mann and Lazier 1996). The sun’s radiation that reaches the sea surface, known as short-wave radiation, is absorbed as it passes through the water column and the intensity at each wavelength decreases exponentially with depth (fig. 2). Figure 2, however, represents the water column as if it was completely clear. There are other factors in the field, such as chlorophyll-a, turbidity, and sedimentation can make the ocean water less clear. The Beer-Lambert Law is an equation that calculates the intensity of the light at a specific wavelength at depth where I is the light intensity after absorption, I<sub>o</sub> is the initial intensity before absorption, *e* is the exponential decay, *a* is the wavelength, and *z* is the depth.

As demonstrated by Lüning and Dring (1972), blue light is the determining factor to initiate a dormant state or to initiate gametogenesis. During the winter, the gametophyte stage persists because the irradiance is low enough where there isn’t enough blue light to initiate gametogenesis. It is important for gametophytes to receive a specific amount of blue light in the winter to initiate gametogenesis and produce an egg from where the sporophyte grows in the spring/summer (Weigel et al. 2023). Therefore my main question for this analysis is, is the blue light irradiance at depths 5m, 10m, 20m, and 40m statistically different during the winter? My null hypothesis is that the average daily irradiances at each depth (5m, 10m, 20m, and 40m) during winter months (November - February) will be the same. My alternative hypothesis is that the average daily irradiances at each depth (5m, 10m, 20m, and 40m) during winter months (November - February) will be different. 

It is important to enhance our knowledge on gametogenesis because the production of an oogonium is necessary for kelp recruitment, and kelp recruitment is a key process in kelp forest dynamics (Graham et al. 2007, Kaminsky et al. 2024). Vegetative microscopic stages of macroalgae were reported to be an important factor in their recruitment success after a stress event (Edwards 2000, Ladah and Zertuche-González 2007, Carney and Edwards 2010, Oppliger et al. 2012). These environmental cues can impact the phase changes in kelp gametophytes which are integral to establish populations in the future (Muth et al. 2019, García-Reyes et al. 2022). Kelp forests are foundation species that provide energy and habitat for all trophic levels, ecosystem services, and climate change remediation (Steneck et al. 2002, Layton et al. 2020). Therefore, it is vital to understand the mechanisms of phase shifts from a vegetative state to gametogenesis since it could be a major factor in the resilience and recovery of kelp populations after stress events. This knowledge gap is particularly important to address due to the decimation of California bull kelp forests from the 2014 - 2016 marine heatwave and subsequent El Niño that led to a rise in *Strongylocentrotus purpuratus* grazing and increased temperatures that were unsuitable for survival (Rogers-Bennett and Catton 2019, McPherson et al. 2021, Edwards 2022, Arafeh-Dalmau et al. 2023, Korabik et al. 2023). 


### Methods 

In order to complete this analysis, I found historical photosynthetically active radiation (PAR) data from the [California Ocean Observing Systems Data Portal](https://data.caloos.org/#metadata/20363/station/data) at the Humboldt Bay Pier (NDBC: HBXC1)(fig. 3). This website provides real time and historical data. Since my goal is to examine irradiances over the course of a completed winter season, I subsetted the historical data that ranged from 2015 to 2025 to November 1st 2023- to 1st February 2024. These dates are relevant to the reproductive season of bull kelp. The data contains columns of time, surface downwelling shortwave flux, and depth (0m). The sensor records measurements every 15 minutes, therefore there are 96 measurements in one day. A large number of measurements per day over the course of three months results in a time series graph with a lot of noise (fig. 4). Therefore, I calculated the daily average irradiances so there would be one value for each day and the data would be more smooth. Since this sensor only provides the shortwave flux at the surface, I used the Beer-Lambert equation to find the irradiances at each depth (fig. 5).

### Statistical Analyses
To test whether the blue light irradiances at 5m, 10m, 20m, and 40m were significantly different over the course of three months, I performed repeated measures ANOVA. This type of test suits this analysis because the irradiance measurements are repeated over time and the independent variable has categories that are related groups, where each group is a specific time point, and is used to detect any overall differences between related means (Laerd Statistics). This test assumes normality and sphericity. While repeated measurements indicate a correlation between observations, it violates the independence assumptions in a regular ANOVA which leads to biased analysis results and incorrect interpretations. However, repeated measures ANOVA accounts for the correlation within and between experimental groups along with the time of the measurements (Muhammed 2023). Sphericity is where the variances of the differences between all combinations of related groups are equal (Pingouin-Stat). To test for normality, I created frequency histograms of irradiance (PAR) at each depth. Figure 6 reveals a fairly normal distribution of average daily irradiance measurements at each depth. 

```python
#Plot the daily average irradiances at each depth as a frequency histogram 
plt.figure(figsize=(9, 7))
plt.suptitle('Histograms of Frequency')

plt.subplot(2, 2, 1)
plt.hist(average_daily_irradiances['irradiance_at_5m'], bins=50, color='red', edgecolor='black', alpha=0.5, label='5m')
plt.gca().set_xticklabels([])
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(average_daily_irradiances['irradiance_at_10m'], bins=50, color='green', edgecolor='black', alpha=0.5, label='10m')
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(average_daily_irradiances['irradiance_at_20m'], bins=50, color='blue', edgecolor='black', alpha=0.5, label='20m')
plt.xlabel("PAR ($\mu$mol m$^{-2}$ s$^{-1}$)")
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 4)
plt.hist(average_daily_irradiances['irradiance_at_40m'], bins=50, color='yellow', edgecolor='black', alpha=0.5, label='40m')
plt.gca().set_yticklabels([])
plt.xlabel("PAR ($\mu$mol m$^{-2}$ s$^{-1}$)")
plt.legend()

plt.tight_layout()
```

### Results 

The results of the repeated measures ANOVA reveals that the data violates the sphericity assumptions which means that the variances of the differences between all combinations of related groups are not equal. This violation increases the change of committing a Type I error (Laerd Statistics). The Mauchley’s Test of Sphericity proves the data violates sphericity because the W-spher and p-sher values that are close to 0. However, the Green-Geisser correction combats the violation of the assumption of sphericity by reducing the degrees of freedom of the F-statistic. By decreasing the degrees of freedom, it requires a larger F-value to achieve statistical significance (Blanca et al. 2023). Therefore, when examining the results, the p-GG-corr is the p-value that determines whether to accept or reject the null hypothesis with an alpha level of 0.05. The p-GG-corr value is 7.40E-35 which is less than 0.05, therefore I reject the null hypothesis and accept the alternative hypothesis. Additionally, the large F-statistic value indicates highly significant differences in the mean of average daily irradiances across different depths. Overall, this test indicates that the irradiances between groups are statistically significant. To further the analysis, I performed a post-hoc paired T-test to examine the differences in average daily irradiance between various depths. To control for multiple comparisons, I used a Bonferroni correction (Armstrong 2014). The pairwise comparisons between the different depth levels showed statistically significant differences (p-corr =7.401010e-34) which is extremely low and indicates very strong evidence for statistical differences. The T-statistic (T statistic = (sample mean - true mean)/ standard error) provides insight into the validity of the Beer-Lambert model. If the corresponding values in column A are less than the values in column B, such as 10m vs 20m, 10m vs. 40m, and 20m vs. 40m, the T-statistic value is positive. If the corresponding values in column A are greater than its pair in column B, such as 10m vs. 5m, 20m vs. 5m, and 40m vs. 5m, the T-statistic is negative. The T-statistic values prove the Beer-Lampert Law is true, where shallower depths will have higher irradiances than deeper depths.

```python
#In order to perform a repeated measures ANOVA, the dataset must be in a long format not wide format

#Identify the columns that represent the repeated measures (irradiance at different depths)
irradiance_columns = ['surface_downwelling_photosynthetic_photon_flux_in_air', 'irradiance_at_5m', 'irradiance_at_10m', 'irradiance_at_20m', 'irradiance_at_40m']

#Created a column called 'time' to recognize the time values in the new long data format
average_daily_irradiances['time'] = average_daily_irradiances.index.to_series()

#Use the melt function to reshape the average daily irradiance data into long format
df_long = pd.melt(average_daily_irradiances,
                    id_vars=['time'], 
                    value_vars=irradiance_columns,
                    var_name='depth',
                    value_name='irradiance')

# Check for NaN values
print("Nan values:",df_long['irradiance'].isnull().sum())

#Convert 'depth' column to a numeric value. This is important for the long format
df_long['irradiance'] = pd.to_numeric(df_long['irradiance'], errors='coerce') 

#Perform the repeated measures ANOVA
rm_anova = pg.rm_anova(data=df_long, dv='irradiance', within='depth', subject='time')

#Print the ANOVA table
print("Repeated Measures ANOVA Table:",rm_anova)
```

```python
#Apply the Bonferroni Correction
posthoc = pg.pairwise_ttests(data=df_long, dv='irradiance', within='depth', subject='time', padjust='bonf')

#Print the posthoc table
print("Posthoc Table:", posthoc)
```

### Conclusion 

Given the p-value in the repeated measures ANOVA was less than 0.05, I reject the null hypothesis and accept the alternative hypothesis that the average daily irradiances at each depth (5m, 10m, 20m, and 40m) during winter months (November - February) are statistically different. The post hoc test confirmed that the daily average irradiances between each depth were statistically significant as well. These results have implications for bull kelp recruitment. The analyses revealed that the amount of blue light available at each depth is different, therefore it may take different amounts of blue light at each depth in order to trigger gametogenesis. This implies that it might take longer for gametogenesis to occur at deeper depths than at shallower depths. This information also contributes to determining the depth range for bull kelp growth. For instance, one reason for why bull kelp does not grow at depths <5m could be because the irradiance is too high for gametophytes to survive and will bleach biological tissue. Conversely, the irradiance may be too low to initiate gametogenesis and sustain growth at depths >30m. In a future study, I can use this data to find the lowest irradiances during the winter and apply those PAR values to laboratory experiments and test for the amount of blue light needed to trigger gametogenesis. This information would enhance knowledge of bull kelp recruitment because the specific blue light threshold at which dormancy shifts to gametogenesis is not well known in bull kelp (Vadas 1972).

### Future Work

In this analysis, I used the Beer-Lambert Law to model blue light irradiances at different depths. However there are some drawbacks to using a model as opposed to collecting real-time data. The model predicts an exponential attenuation of light with depth, however there are factors that are not accounted for that would impact results of PAR attenuation in the field, such as chlorophyll a, turbidity, and sedimentation (Aumack et al. 2007, Weiskerger et al. 2018). Shading from canopy and light flecking can also impact the amount of light available at the benthos and can limit benthic vegetation productivity (Connell et. al 2005, Graham et al. 2007, Aumack et al. 2007, Pedersen et al. 2012). Weiskerger et al. 2018 found that measuring other variables, such as organic particles, inorganic particles, and dissolved particles, were reliable predictors of PAR using the Beer-Lambert equation where the model captured over 90% of the variability in PAR data over the water column. Therefore, it would be beneficial to measure PAR  at different depths over a period of time using PAR sensors deployed at each depth and measure other variables such as particle size, shape, and composition that could influence scattering and attenuation in a future study. Comparing the in-situ data and the modeled data could provide insight into the accuracy of the model and the importance of light for biological processes, such as the initiation of gametogenesis in bull kelp gametophytes.
