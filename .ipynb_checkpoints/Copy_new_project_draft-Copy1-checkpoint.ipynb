### Comparing the blue light available at different depths (m)
 This notebook uses historical surface PAR data from the Humboldt Bay Pier (NDBC: HBXC1) to calculate light attenuation at different depths where bull kelp gametophytes could settle, fertilize, and grow. The historical data needed for this analysis is found on the [California Ocean Observing Systems Data Portal](https://data.caloos.org/#metadata/20363/station/data).

```python
df = pd.read_csv('humboldt_PAR.csv')
df['time'] = pd.to_datetime(df['time'], errors='coerce', utc=True)

print(df[df['time'].isnull()])

start_date = pd.to_datetime('2023-11-01T00:00:00Z')
end_date = pd.to_datetime('2024-02-01T00:00:00Z')

subset = (df['time'] >= start_date) & (df['time'] <= end_date)

df2 = df.loc[subset, ['time', 'surface_downwelling_photosynthetic_photon_flux_in_air']]

new_df = df2.set_index('time') #time is now the index to organize the data by the time not the sequence of numbers
print(new_df.index)
```

### Analyzing the data from the new data file 

```python
#Importing function
import sys
sys.path.append('/Users/nataliecolao@Natalies-MacBook-Pro-2/desktop/python/Final_project_data/draft_project')
import beer_lampert_law as bll

#The next line removes the NaN values from the dataset
cleaned_data = new_df.dropna(subset=['surface_downwelling_photosynthetic_photon_flux_in_air'])

surface_irradiance = cleaned_data['surface_downwelling_photosynthetic_photon_flux_in_air']
depths = [5.0, 10.0, 20.0, 40.0]
attenuation_coefficient = 0.004 #this value is the blue light wavelength

for depth in depths:
    new_column = f'irradiance_at_{int(depth)}m' #this is creating new columns with the irradiance at each depth 
    cleaned_data[new_column] = bll.beer_lampert_law(surface_irradiance, attenuation_coefficient, depth) #this applies the function to the dataset without NaN values
print(cleaned_data[new_column])
```

```python
average_daily_irradiances = cleaned_data.resample("D").mean()
print(average_daily_irradiances['irradiance_at_10m'])

plt.figure()
plt.plot(average_daily_irradiances.index, average_daily_irradiances['surface_downwelling_photosynthetic_photon_flux_in_air'], 'r-', label='surface 0m')
plt.xlabel('Date')
plt.ylabel('PAR ($\mu$mol m$^{-2}$ s$^{-1}$)')
plt.ylim(0,250)
plt.title('Surface Downwelling PAR in Humboldt from November 2023 to March 2024 at the surface 0m')
plt.xticks(rotation=90)
plt.legend()
```

```python
import pandas as pd

# Identify the columns that represent your repeated measures (irradiance at different depths)
irradiance_columns =['surface_downwelling_photosynthetic_photon_flux_in_air', 'irradiance_at_5m', 'irradiance_at_10m', 'irradiance_at_20m', 'irradiance_at_40m']

# Use the melt function to reshape the DataFrame
df_long = pd.melt(average_daily_irradiances,
                  id_vars= ['time'],  # The column(s) that identify each observation
                  value_vars=irradiance_columns,  # The columns to unpivot
                  var_name='depth',  # The name for the new column containing the former column names
                  value_name='irradiance')  # The name for the new column containing the values

# Now 'df_long' will be in long format, looking something like this:
#
#      time                 depth  irradiance
# 0     0.0  surface_irradiance         100
# 1     0.5  surface_irradiance         110
# 2     1.0  surface_irradiance         120
# 3     0.0       irradiance_5m          90
# 4     0.5       irradiance_5m          95
# 5     1.0       irradiance_5m         100
# ...

# You might want to clean up the 'depth' column by removing the 'irradiance_' prefix
df_long['depth'] = df_long['depth'].str.replace('irradiance_', '')


# It's a good idea to check for any NaN values that might have been introduced
# by the conversion (if there were non-numeric values in the original column)
print(df_long['irradiance'].isnull().sum())
```

```python
import pingouin as pg

# Assuming your long format DataFrame is named 'df_long'
# and has columns 'time', 'depth', and 'irradiance'

df_long['irradiance'] = pd.to_numeric(df_long['irradiance'], errors='coerce')

# Perform the repeated measures ANOVA
aov = pg.rm_anova(data=df_long, dv='irradiance', within='depth', subject='time')

# Print the ANOVA table
print(aov)
```

```python
#Bonferroni Correction
posthoc = pg.pairwise_ttests(data=df_long, dv='irradiance', within='depth', subject='time', padjust='bonf')
print(posthoc)
```

Based on the repeated measures ANOVA and the post-hoc pairwise comparisons, you can conclude that there is a significant effect of depth on irradiance, and furthermore, the irradiance levels are significantly different between every pair of the measured depths (surface, 5m, 10m, 20m, and 40m). The negative t-values for comparisons where 'surface_downwelling_photosynthetic_photon_flux_in_air' or shallower depths are compared to deeper depths (e.g., 'at_5m' vs. 'at_40m') suggest that irradiance tends to be higher at shallower depths. Conversely, positive t-values (e.g., 'at_10m' vs. 'at_20m') suggest that irradiance is higher at the first depth listed in the pair.

Since the irradiances at each depth are statistically sig, this indiactes that there could be different effects at each depth on gametogeneis

```python
def calculate_daily_average(data):
    """
    Used for time series frequency conversion and averaging. The argument "D" is a frequency string that signifies "calendar day" and calculates mean of the numerical columns for each day.

    Parameters
    ------------
    data : pd.DataFrame
        This is the time series data (Pandas DataFrame with a DatetimeIndex)
        that will be converted to a calendar day and averaged.

    Return
    ------------
    This is a new DataFrame containing the averaged values for each
        calendar day. The index will be the daily dates.

    Args:
        data : pd.DataFrame
            Time series data with a DatetimeIndex.

    Returns:
        pd.DataFrame: DataFrame with daily average values.
        This is a new DataFrame containing the averaged values for each
        calendar day and the index will be the daily dates. 
    """
    return data.resample("D").mean()

average_daily_irradiances = calculate_daily_average(cleaned_data)
print(average_daily_irradiances)
```
