```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('humboldt_radiation.csv')
```

```python
df = pd.read_csv('humboldt_radiation.csv')
df['time'] = pd.to_datetime(df['time'], errors='coerce', utc=True)

print(df[df['time'].isnull()])

start_date = pd.to_datetime('2023-11-01T00:00:00Z')
end_date = pd.to_datetime('2024-05-01T00:00:00Z')

subset = (df['time'] >= start_date) & (df['time'] <= end_date)

new_df = df.loc[subset, ['time', 'surface_downwelling_shortwave_flux_in_air']]

print(new_df)
```

```python
plt.figure()
plt.plot(new_df['time'], new_df['surface_downwelling_shortwave_flux_in_air'], '-')
plt.ylim(0, 160)
plt.xlabel('Date')
plt.ylabel('Surface Downwelling Shortwave Flux in Air (W/m$^2$)')
plt.title('Surface Downwelling Shortwave Flux in Humboldt from November 2023 to May 2024')
```
